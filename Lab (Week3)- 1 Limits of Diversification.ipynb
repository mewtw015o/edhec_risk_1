```python
total_market_index = erk.drawdown(total_market_return).Wealth
total_market_index.plot(figsize=(12,6),title="Total Market CapWeighted Index 1926-2018")
```

```python
total_market_index["1980":].plot(figsize=(12,6))
#Moving Average
total_market_index["1980":].rolling(window=36).mean().plot(figsize=(12,6)) #rolling return of 3 years or 36 months by their mean
```

```python
tmi_tr36rets = total_market_return.rolling(window=36).aggregate(erk.annualized_rets,periods_per_year=12) #compounding returns in each window
tmi_tr36rets.plot(figsize=(12,6),label="Tr36 mo Return",legend=True)
total_market_return.plot(label="Returns",legend=True)
```

### Rolling Correlation - along with MultiIndexes and .groupby

```python
ts_corr = ind_return.rolling(window=36).corr() #rolling behaves like groupby - grouping the last 36 rows
ts_corr.tail() #time series of matrix; multi-index (date, industry)
```

```python
ts_corr.index.names
```

```python
ts_corr.index.names = ['date','industry'] #renaming multi-index
```

```python
ts_corr.tail()
```

```python
#Trailing 36 months average correlation
ind_tr36corr = ts_corr.groupby(level='date').apply(lambda cormat : cormat.values.mean())
ind_tr36corr.plot()
```

```python
tmi_tr36rets.plot(label="Tr36 Mo Rets",figsize=(12,6), legend=True)
ind_tr36corr.plot(label="Tr36 Mo Corr",legend=True)
```

```python
tmi_tr36rets.plot(label="Tr36 Mo Rets",figsize=(12,6), legend=True)
ind_tr36corr.plot(label="Tr36 Mo Corr",legend=True, secondary_y=True)
```
